**Reviews and Ratings of the restaurants in the most popular city**

```python
ecity_rest=df[df['Area']=='Electronic City, Bangalore']


ecity_rest.loc[:, 'Dinner Ratings'] = pd.to_numeric(
    ecity_rest['Dinner Ratings'], errors='coerce')
ecity_rest.loc[:, 'Dinner Reviews'] = pd.to_numeric(
    ecity_rest['Dinner Reviews'], errors='coerce')

# Sort the filtered DataFrame by 'Dinner Ratings' in descending order
top_rated_electronic_city_restaurants = ecity_rest.sort_values(
    by='Dinner Ratings', ascending=False)

# Then, sort by 'Dinner Reviews' in descending order
top_rated_electronic_city_restaurants = top_rated_electronic_city_restaurants.sort_values(
    by='Dinner Reviews', ascending=False)

# Display the top-rated restaurants sorted by dinner reviews
print(top_rated_electronic_city_restaurants[['Name', 'Dinner Ratings', 'Dinner Reviews']].head(10))
```

```python
top_rest_rating=top_rated_electronic_city_restaurants[top_rated_electronic_city_restaurants]
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Create a bar plot
plt.figure(figsize=(10, 6))
top_rated_electronic_city_restaurants[['Name', 'Dinner Ratings']][:15].plot(kind='bar')
plt.title('Preferred restaurants  in Electronic City Banglore')
plt.xlabel('Restaurant')
plt.ylabel('Ratings')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```
